```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

df = pd.read_excel("Customer_Churn_Dataset.xlsx")
print("Shape:", df.shape)
print("\nColumns:\n", df.columns.tolist())
print("\nData sample:")
display(df.head())
print("\nInfo:")
df.info()
print("\nMissing values per column:")
print(df.isna().sum())
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['MonthlyCharges'] = pd.to_numeric(df['MonthlyCharges'], errors = 'coerce')
df['Tenure'] = pd.to_numeric(df['Tenure'], errors = 'coerce')

before = df.shape[0]
df = df.drop_duplicates(subset = ['CustomerID'])
after = df.shape[0]
print(f"Removed {before-after} duplicates rows (if any).")

bad_total = df[df['TotalCharges'].isna()]
print ("Rows with NaN TotalCharges:",bad_total.shape[0])
display(bad_total.head())
```

```python
mask = df['TotalCharges'].isna() & df['Tenure'].notna() & df['MonthlyCharges'].notna()
df.loc[mask,'TotalCharges'] = (df.loc[mask,'Tenure'] * df.loc[mask,'MonthlyCharges']).round(2)

num_clos = ['MonthlyCharges','Tenure','TotalCharges']
for col in num_clos:
    if df[col].isna().sum() > 0:
        df[col] = df[col].fillna(df[col].median())

cat_cols = df.select_dtypes(include='object').columns.tolist()
for col in cat_cols:
    df[col] = df[col].fillna('Unknown')

print("Missing values after cleaning:")
print(df.isna().sum())
```

```python
df['ChurnFlag'] = df['Churn'].apply(lambda x: 1 if str(x).strip().lower() == 'yes' else 0)

bins = [-1,6,12,24,48,1000]
labels = ['0-6','7-12','13-24','25-48','49+']
df['TenureGroup'] = pd.cut(df['Tenure'], bins=bins, labels=labels)

df[['CustomerID','Tenure','TenureGroup','Churn','ChurnFlag']].head()
```

```python
total_customers = df.shape[0]
churned = df['ChurnFlag'].sum()
churn_rate = churned/total_customers

print(f"Total customers: {total_customers}")
print(f"Churned customer: {churned}")
print(f"Overall churn rate: {churn_rate:.2%}")
```

```python
churn_by_contract = df.groupby('Contract')['ChurnFlag'].mean().sort_values(ascending=False)
churn_by_internet = df.groupby('InternetService')['ChurnFlag'].mean().sort_values(ascending=False)
churn_by_tenure = df.groupby('TenureGroup')['ChurnFlag'].mean().sort_values(ascending=False)

print("\nChurn rate by Contract:")
print((churn_by_contract * 100).round(2).astype(str) + '%')

print("\nChurn rate by Internet Service:")
print((churn_by_internet * 100).round(2).astype(str) + '%')

print("\nChurn rate by Tenure Group:")
print((churn_by_tenure * 100).round(2).astype(str) + '%')
```

```python
import os
os.makedirs("charts",exist_ok=True)

(df.groupby("Contract")["ChurnFlag"].mean().sort_values(ascending=False).plot(kind="bar",title="Churn Rate by Contract"))
plt.ylabel("Churn Rate")
plt.tight_layout()
plt.savefig("charts/churn_by_contract.png")
plt.show()
plt.clf()

(df.groupby("InternetService")["ChurnFlag"].mean().sort_values(ascending=False).plot(kind="bar",title="Churn Rate by Internet Service"))
plt.ylabel("Churn Rate")
plt.tight_layout()
plt.savefig("charts/churn_by_internet.png")
plt.show()
plt.clf()

(df.groupby("TenureGroup")["ChurnFlag"].mean().sort_values(ascending=False).plot(kind="bar",title="Churn Rate by Tenure Group"))
plt.ylabel("Churn Rate")
plt.tight_layout()
plt.savefig("charts/churn_by_tenure.png")
plt.show()
plt.clf()
```

```python
summary = f"""
 Customer Churn Analysis  Key Insights

1 Overall Churn Rate:
- The overall churn rate is {(df['ChurnFlag'].mean() * 100):.2f}%, which indicates almost half of customers are leaving.
- This is significantly high, suggesting urgent retention action.

2 High-Risk Contract Type:
- One-year contract customers show the highest churn rate, followed by Month-to-month.
- Long-term plan customers (Two-year) churn less.
 Customers feel more satisfied or locked in long-term plans.

3 Internet Service Risk Pattern:
- Customers with No internet service churn the most, which implies:
  - These customers depend only on phone services.
  - Highly competitive alternatives may drive them away.

4 Tenure Group  When customers leave
- Highest churn occurs in the 712 month group.
- This means many customers leave within their first year, possibly after:
  - Trial period ends  
  - Price increases  
  - Dissatisfaction during onboarding

5 Customer Behavior Pattern
- Short-tenure customers (06 months) also show high churn  early dissatisfaction.
- Stable customers (25+ months) tend to stay longer  loyal base.

 Recommendation Summary
- Improve onboarding experience for new customers (first 612 months).
- Offer renewal incentives for 1-year contract customers.
- Bundle internet service for customers without internet to reduce churn.
- Track satisfaction for customers during month 712  highest drop-off point.
"""

print(summary)
```

```python
from reportlab.lib.pagesizes import letter
from reportlab.lib.units import inch
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Image, PageBreak, KeepTogether, ListFlowable, ListItem
import os, textwrap

output_pdf = "customer_churn_report.pdf"
charts = ["charts/churn_by_contract.png", "charts/churn_by_internet.png", "charts/churn_by_tenure.png"]

clean_summary = summary.encode("ascii", "ignore").decode().strip()
clean_summary = clean_summary.replace("712", "712").replace("06", "06").replace("612", "612")
paras = [p for p in clean_summary.split("\n") if p.strip()]
if len(paras) == 1:
    wrapped = textwrap.fill(paras[0], width=110)
    paras = wrapped.split("\n\n") if "\n\n" in wrapped else wrapped.split("\n")

doc = SimpleDocTemplate(output_pdf, pagesize=letter,
                        leftMargin=0.7*inch, rightMargin=0.7*inch,
                        topMargin=0.6*inch, bottomMargin=0.6*inch)

styles = getSampleStyleSheet()
title_style = ParagraphStyle('Title', parent=styles['Heading1'], fontSize=26, leading=30, alignment=1, spaceAfter=12)
subtitle_style = ParagraphStyle('Subtitle', parent=styles['Heading2'], fontSize=12, leading=14, alignment=1, spaceAfter=8)
heading_style = ParagraphStyle('Heading', parent=styles['Heading2'], fontSize=14, leading=18, spaceBefore=8, spaceAfter=6)
body_style = ParagraphStyle('Body', parent=styles['BodyText'], fontSize=11, leading=15)
bullet_style = ParagraphStyle('Bullet', parent=styles['BodyText'], fontSize=11, leading=14, leftIndent=12)

your_name = "Aniket Singh"
your_role = "Data Analyst | Python, Pandas, Matplotlib"

flow = []
flow.append(Spacer(1, 1.0*inch))
flow.append(Paragraph("Customer Churn Analysis", title_style))
flow.append(Paragraph("Telecom Customer Churn Dataset", subtitle_style))
flow.append(Spacer(1, 0.2*inch))
flow.append(Paragraph(f"Author: <b>{your_name}</b>", body_style))
flow.append(Paragraph(f"Role: {your_role}", body_style))
flow.append(Spacer(1, 0.25*inch))
flow.append(Paragraph("<b>Tools used:</b> Python, pandas, matplotlib", body_style))
flow.append(Paragraph("<b>Deliverables:</b> Data cleaning, EDA, Churn insights, Charts, PDF report", body_style))
flow.append(PageBreak())

flow.append(Paragraph("Executive Summary", heading_style))
for p in paras:
    wrapped = textwrap.fill(p, width=110)
    flow.append(Paragraph(wrapped.replace("\n", "<br/>"), body_style))
    flow.append(Spacer(1, 0.12*inch))
flow.append(PageBreak())

flow.append(Paragraph("Key Findings & Recommendations", heading_style))
total_customers = df.shape[0]
churn_rate_pct = df['ChurnFlag'].mean() * 100
bullets = [f"Overall churn rate: {churn_rate_pct:.2f}% ({total_customers} customers)."]
churn_by_contract = df.groupby('Contract')['ChurnFlag'].mean().sort_values(ascending=False)
if not churn_by_contract.empty:
    bullets.append(f"Highest churn by contract: {churn_by_contract.index[0]} (highest churn rate).")
churn_by_internet = df.groupby('InternetService')['ChurnFlag'].mean().sort_values(ascending=False)
if not churn_by_internet.empty:
    bullets.append(f"Highest churn by internet service: {churn_by_internet.index[0]}.")
bullets.extend([
    "Focus retention efforts in months 612 (onboarding + renewals).",
    "Offer bundle incentives to customers without internet.",
    "Monitor one-year contract renewals and offer targeted discounts."
])

list_items = [ListItem(Paragraph(b, bullet_style)) for b in bullets]
flow.append(ListFlowable(list_items, bulletType='bullet'))
flow.append(PageBreak())

for fname in charts:
    if os.path.exists(fname):
        img = Image(fname)
        max_width = 7.8 * inch
        if img.drawWidth > max_width:
            ratio = max_width / float(img.drawWidth)
            img.drawWidth = max_width
            img.drawHeight = float(img.drawHeight) * ratio
        flow.append(KeepTogether([Spacer(1, 0.8*inch), img, Spacer(1, 0.15*inch),
                                 Paragraph(os.path.basename(fname).replace("_"," ").replace(".png","").title(), subtitle_style)]))
        flow.append(PageBreak())
    else:
        flow.append(Paragraph(f"Missing chart file: {fname}", body_style))
        flow.append(PageBreak())

flow.append(Paragraph("This Report generated with Python (pandas, matplotlib).", body_style))

doc.build(flow)
print("Premium bullet-style PDF created at:", os.path.abspath(output_pdf))
```
